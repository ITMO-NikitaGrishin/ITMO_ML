#### **$z < 0$**

#### **$0 \leq z \leq 1$**
–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è: –æ—Ç $x = 0$ –¥–æ $x = z$

$$
= 2 \left[ -\frac{1}{2}e^{-2x} \right]_{0}^{z} = -[e^{-2z} - 1] = 1 - e^{-2z}
$$

#### **$z > 1$**
–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è: –æ—Ç $x = s-1$ –¥–æ $x = z$

$$
= 2 \left[ -\frac{1}{2}e^{-2x} \right]_{s-1}^{z} = -[e^{-2z} - e^{-2(z-1)}]
$$

$$
= e^{-2(z-1)} - e^{-2z} = e^{-2z}(e^{2} - 1)
$$
